```python
import os
import json
import time
from datetime import datetime
from typing import List, Dict, Any

# For LLM interaction - using OpenAI
try:
    from openai import OpenAI
except ImportError:
    print("Installing openai...")
    import subprocess
    subprocess.check_call(['pip', 'install', 'openai'])
    from openai import OpenAI

# Initialize client
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def call_llm(prompt: str, model: str = "gpt-3.5-turbo", temperature: float = 0.7, max_tokens: int = 2000) -> str:
    """Call OpenAI API with given prompt"""
    try:
        response = client.chat.completions.create(
            model=model,
            messages=[{"role": "user", "content": prompt}],
            temperature=temperature,
            max_tokens=max_tokens
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"Error: {str(e)}"

print("Setup complete. LLM functions ready.")
```

```python
# INTERVIEW APPROACH
# Simulating an interview where the model is asked follow-up questions

def interview_approach(topic: str) -> Dict[str, str]:
    """Interview approach: Ask follow-up questions to get deeper insights"""
    
    interview_results = {}
    
    # Initial question
    initial_prompt = f"I want to learn about {topic}. Start by giving a brief overview."
    initial_response = call_llm(initial_prompt, temperature=0.7)
    interview_results['initial_overview'] = initial_response
    
    print(f"Q1 (Overview): {initial_prompt}\n")
    print(f"A1:\n{initial_response}\n")
    print("-" * 80 + "\n")
    
    # Follow-up question 1
    followup1 = f"Based on your previous answer about {topic}, what are the key challenges or limitations?"
    response1 = call_llm(followup1, temperature=0.7)
    interview_results['challenges'] = response1
    
    print(f"Q2 (Challenges): {followup1}\n")
    print(f"A2:\n{response1}\n")
    print("-" * 80 + "\n")
    
    # Follow-up question 2
    followup2 = f"What are the practical applications of {topic} in real-world scenarios?"
    response2 = call_llm(followup2, temperature=0.7)
    interview_results['applications'] = response2
    
    print(f"Q3 (Applications): {followup2}\n")
    print(f"A3:\n{response2}\n")
    print("-" * 80 + "\n")
    
    # Follow-up question 3
    followup3 = f"How might {topic} evolve in the future?"
    response3 = call_llm(followup3, temperature=0.7)
    interview_results['future'] = response3
    
    print(f"Q4 (Future): {followup3}\n")
    print(f"A4:\n{response3}\n")
    
    return interview_results

# Example usage
print("=" * 80)
print("INTERVIEW APPROACH DEMONSTRATION")
print("=" * 80 + "\n")

topic = "Artificial Intelligence"
interview_data = interview_approach(topic)
```

```python
# CHAIN OF THOUGHT (COT) APPROACH
# Prompting the model to think step-by-step before giving an answer

def chain_of_thought(problem: str) -> Dict[str, str]:
    """Chain of Thought: Ask model to reason step-by-step"""
    
    cot_prompt = f"""
Solve the following problem step by step. Show your reasoning at each step.

Problem: {problem}

Please:
1. Break down the problem
2. Identify the key components
3. Work through the logic step-by-step
4. Arrive at the final answer
"""
    
    response = call_llm(cot_prompt, temperature=0.7, max_tokens=2000)
    
    print("=" * 80)
    print("CHAIN OF THOUGHT (COT) APPROACH")
    print("=" * 80 + "\n")
    print(f"Problem: {problem}\n")
    print("CoT Response:")
    print(response)
    print("-" * 80 + "\n")
    
    return {'problem': problem, 'cot_response': response}

# Example usage
problem1 = "A store sells apples at $2 each. If you buy 5 apples and get a 10% discount, how much do you pay?"
cot_result = chain_of_thought(problem1)
```

```python
# TREE OF THOUGHT (TOT) APPROACH
# Exploring multiple reasoning paths and selecting the best one

def tree_of_thought(problem: str, num_branches: int = 3) -> Dict[str, Any]:
    """Tree of Thought: Generate multiple reasoning paths and evaluate them"""
    
    print("=" * 80)
    print("TREE OF THOUGHT (TOT) APPROACH")
    print("=" * 80 + "\n")
    print(f"Problem: {problem}\n")
    
    # Step 1: Generate multiple approaches
    generation_prompt = f"""
For the problem: {problem}

Generate {num_branches} different approaches to solve this problem. 
Be concise and label them as Approach 1, Approach 2, Approach {num_branches}.
"""
    
    approaches_response = call_llm(generation_prompt, temperature=0.9, max_tokens=1500)
    print("Generated Approaches:")
    print(approaches_response)
    print("-" * 80 + "\n")
    
    # Step 2: Evaluate each approach
    evaluation_prompt = f"""
Evaluate the following approaches for solving: {problem}

{approaches_response}

For each approach, rate it on:
1. Correctness (0-10)
2. Efficiency (0-10)
3. Feasibility (0-10)

Provide ratings and brief justification for each.
"""
    
    evaluation_response = call_llm(evaluation_prompt, temperature=0.7, max_tokens=1500)
    print("Evaluation of Approaches:")
    print(evaluation_response)
    print("-" * 80 + "\n")
    
    # Step 3: Select and refine best approach
    refinement_prompt = f"""
Based on the evaluation above, which approach is best for solving: {problem}

Please provide the refined, detailed solution using the best approach.
"""
    
    final_response = call_llm(refinement_prompt, temperature=0.7, max_tokens=1500)
    print("Final Refined Solution (Best Approach):")
    print(final_response)
    print("-" * 80 + "\n")
    
    return {
        'problem': problem,
        'approaches': approaches_response,
        'evaluation': evaluation_response,
        'final_solution': final_response
    }

# Example usage
problem2 = "Design a recommendation system for an e-commerce platform"
tot_result = tree_of_thought(problem2, num_branches=3)
```

```python
# ZERO-SHOT PROMPTING
# No examples provided - model uses only generic knowledge

def zero_shot_prompting(task: str, input_text: str) -> str:
    """Zero-shot: Direct task without examples"""
    
    prompt = f"""
Task: {task}

Input: {input_text}

Please perform the task above.
"""
    
    response = call_llm(prompt, temperature=0.7)
    return response

# ZERO-SHOT EXAMPLES
print("=" * 80)
print("ZERO-SHOT PROMPTING (NO EXAMPLES)")
print("=" * 80 + "\n")

# Example 1: Sentiment Analysis (Zero-shot)
zero_shot_sentiment_prompt = "Classify the sentiment of the following text as positive, negative, or neutral"
zero_shot_sentiment_text = "I absolutely love this product! It's amazing and exceeded all my expectations."
zero_shot_sentiment_result = zero_shot_prompting(zero_shot_sentiment_prompt, zero_shot_sentiment_text)

print(f"Task: {zero_shot_sentiment_prompt}")
print(f"Input: {zero_shot_sentiment_text}")
print(f"Zero-shot Result: {zero_shot_sentiment_result}\n")
print("-" * 80 + "\n")

# Example 2: Named Entity Recognition (Zero-shot)
zero_shot_ner_prompt = "Identify and extract all person names and organizations from the following text"
zero_shot_ner_text = "Apple Inc. was founded by Steve Jobs, Bill Gates founded Microsoft, and both are tech pioneers."
zero_shot_ner_result = zero_shot_prompting(zero_shot_ner_prompt, zero_shot_ner_text)

print(f"Task: {zero_shot_ner_prompt}")
print(f"Input: {zero_shot_ner_text}")
print(f"Zero-shot Result:\n{zero_shot_ner_result}\n")
print("-" * 80 + "\n")

# Example 3: Text Classification (Zero-shot)
zero_shot_class_prompt = "Classify the following news headline as breaking news, feature, or analysis"
zero_shot_class_text = "New breakthrough in quantum computing announced by researchers"
zero_shot_class_result = zero_shot_prompting(zero_shot_class_prompt, zero_shot_class_text)

print(f"Task: {zero_shot_class_prompt}")
print(f"Input: {zero_shot_class_text}")
print(f"Zero-shot Result: {zero_shot_class_result}\n")
print("-" * 80 + "\n")

zero_shot_results = {
    'sentiment': zero_shot_sentiment_result,
    'ner': zero_shot_ner_result,
    'classification': zero_shot_class_result
}
```

```python
# FEW-SHOT PROMPTING
# Providing examples to guide the model's behavior

def few_shot_prompting(task: str, examples: List[Dict[str, str]], input_text: str) -> str:
    """Few-shot: Provide examples before asking for the task"""
    
    prompt = f"Task: {task}\n\nExamples:\n"
    
    for i, example in enumerate(examples, 1):
        prompt += f"\nExample {i}:\nInput: {example['input']}\nOutput: {example['output']}"
    
    prompt += f"\n\nNow perform the same task for:\nInput: {input_text}\nOutput:"
    
    response = call_llm(prompt, temperature=0.7)
    return response

# FEW-SHOT EXAMPLES
print("=" * 80)
print("FEW-SHOT PROMPTING (WITH EXAMPLES)")
print("=" * 80 + "\n")

# Example 1: Sentiment Analysis (Few-shot)
sentiment_examples = [
    {
        "input": "I hate this product, it's terrible",
        "output": "Sentiment: Negative"
    },
    {
        "input": "This is okay, nothing special",
        "output": "Sentiment: Neutral"
    },
    {
        "input": "I love this! It's fantastic!",
        "output": "Sentiment: Positive"
    }
]

few_shot_sentiment_prompt = "Classify the sentiment of the text as positive, negative, or neutral"
few_shot_sentiment_text = "Best purchase ever! Highly recommend to everyone!"
few_shot_sentiment_result = few_shot_prompting(few_shot_sentiment_prompt, sentiment_examples, few_shot_sentiment_text)

print(f"Task: {few_shot_sentiment_prompt}")
print("\nExamples provided:")
for i, ex in enumerate(sentiment_examples, 1):
    print(f"  {i}. '{ex['input']}' -> {ex['output']}")
print(f"\nInput: {few_shot_sentiment_text}")
print(f"Few-shot Result: {few_shot_sentiment_result}\n")
print("-" * 80 + "\n")

# Example 2: Question Generation (Few-shot)
qgen_examples = [
    {
        "input": "The Great Wall of China was built over many centuries.",
        "output": "When was the Great Wall of China built?"
    },
    {
        "input": "Photosynthesis is the process by which plants convert sunlight into energy.",
        "output": "What is photosynthesis?"
    },
    {
        "input": "Machine learning is a subset of artificial intelligence.",
        "output": "What is the relationship between machine learning and AI?"
    }
]

few_shot_qgen_prompt = "Generate a question based on the following statement"
few_shot_qgen_text = "Python is a popular programming language for data science."
few_shot_qgen_result = few_shot_prompting(few_shot_qgen_prompt, qgen_examples, few_shot_qgen_text)

print(f"Task: {few_shot_qgen_prompt}")
print("\nExamples provided:")
for i, ex in enumerate(qgen_examples, 1):
    print(f"  {i}. '{ex['input']}' -> {ex['output']}")
print(f"\nInput: {few_shot_qgen_text}")
print(f"Few-shot Result: {few_shot_qgen_result}\n")
print("-" * 80 + "\n")

# Example 3: Text Translation (Few-shot)
translate_examples = [
    {
        "input": "Hello, how are you?",
        "output": "Hola, Â¿cÃ³mo estÃ¡s?"
    },
    {
        "input": "What is your name?",
        "output": "Â¿CuÃ¡l es tu nombre?"
    },
    {
        "input": "Thank you very much",
        "output": "Muchas gracias"
    }
]

few_shot_translate_prompt = "Translate the following English text to Spanish"
few_shot_translate_text = "Good morning, nice to meet you"
few_shot_translate_result = few_shot_prompting(few_shot_translate_prompt, translate_examples, few_shot_translate_text)

print(f"Task: {few_shot_translate_prompt}")
print("\nExamples provided:")
for i, ex in enumerate(translate_examples, 1):
    print(f"  {i}. '{ex['input']}' -> '{ex['output']}'")
print(f"\nInput: {few_shot_translate_text}")
print(f"Few-shot Result: {few_shot_translate_result}\n")
print("-" * 80 + "\n")

few_shot_results = {
    'sentiment': few_shot_sentiment_result,
    'question_generation': few_shot_qgen_result,
    'translation': few_shot_translate_result
}
```
